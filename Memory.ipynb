база<br>

  id - uuid<br>
  url - url как есть в определенном формате(для каждого типа свой)<br>
  source_type — авито циан и тп<br>
  last_check_ts — время последнего чека (nullable)<br>
  is_publicated — опубликовано или снять с публикации(nullable)<br>
  price — цена (nullable)<br>

```python
from sqlalchemy import create_engine
from sqlalchemy.orm import DeclarativeBase
from sqlalchemy.orm import Session
from sqlalchemy import Column, Integer, String, TEXT
from sqlalchemy import select
from sqlalchemy.dialects.postgresql import UUID
import uuid
from sqlalchemy import create_engine
import psycopg2
engine = create_engine("postgresql+psycopg2://admin:admin111@localhost/advertdb")

class Base(DeclarativeBase):
    pass

class Advert(Base):
    __tablename__ = "advert"
    
    id = Column(UUID(as_uuid=True), primary_key=True, default=uuid.uuid4)
    url = Column(TEXT)
    last_check_ts =  Column(TEXT)
    is_publicated = Column(TEXT)
    price = Column(TEXT)
    
    
Base.metadata.create_all(engine)
```
