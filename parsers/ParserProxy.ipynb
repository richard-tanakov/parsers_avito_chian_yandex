```python
from bs4 import BeautifulSoup as bs
from typing import List,Dict
import requests

import json
```

```python
def filtration_proxy(proxs, test_in_active ):
        """ Проверка на актуальность """
        
        url = 'https://www.avito.ru/'
        
                   
        proxies ={
                                        'http' :f'{proxs}',
                                        'https': f'{proxs}',
                                }
        try:
                response = requests.get(url=url,proxies=proxies)
                status = (response.status_code)
                if status == 200:
                    tets_in_active[proxies].append({'p':f'{proxs}'})
                    with open('listproxy/proxy_active.json','a') as f:
                        json.dump(tets_in_active,f)
                else:
                        pass
        except Exception as e:  
            	        pass             
                                            
```

```python
def open_pr_list():
    tets_in_active={}
    tets_in_active['proxies'] = []
    with open ("listproxy/proxy.json", 'r', encoding= 'utf-8' ) as f:
        proxi_list = json.load(f)

        for proxy in proxi_list['proxies']:
                    filtration_proxy(proxy['name'],tets_in_active)
```

```python
def create_proxy_list(created_proxy : Dict):
    """ Парсинг прокси листа """
    created_proxy['proxies'] = []
    for count in range(20):
        
        html = requests.get('https://free-proxy-list.net/')
        print(html.status_code)
        soup = bs(html.text, 'lxml')
        res = soup.find('tbody').find_all('tr')
        
        for i in res:
                if i.find_all('td')[6].text == 'yes':
                    ip = i.find_all('td')[0].text
                    port = i.find_all('td')[1].text
                    proxy_adress =  f'{ip}'+':'+f'{port}'
                    
                    created_proxy['proxies'].append({'name':f'{proxy_adress}'})
                    
    with open('listproxy/proxy.json','w') as f:
                    json.dump(created_proxy,f)
                    
                    
                    
        
```
